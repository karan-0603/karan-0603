```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import LabelEncoder
import gradio as gr
import warnings
import json
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
vectorizer = LabelEncoder()
training_data = pd.read_csv('/content/yoga training dataset - Sheet1.csv')
yoga_poses_data = pd.read_csv('/content/Yoga Poses Data - Sheet1 (2).csv')
training_data['Desired Outcomes'] = vectorizer.fit_transform(training_data['Desired Outcomes'])

# # Separate the features (diseases) and labels (yoga poses)
X = training_data[['Fitness Level','Flexibility Level','Desired Outcomes']]
y = training_data['Pose Id']

# Vectorize the textual data
model = NearestNeighbors(n_neighbors=1)
model.fit(X)

# Find the nearest neighbors based on user input
def getPoses(fitness,flexibility,outcomes):
    outcomes = vectorizer.transform([outcomes.lower()])
    distances, indices = model.kneighbors([[int(fitness),int(flexibility),outcomes[0]]]) # Access the first element of outcomes
    distance = distances[0]
    pose_ids_strings = y[indices[0][0]]
    pose_ids_array = pose_ids_strings.split(",")
    stringData = ""
    count = 0
    result = []
    print(pose_ids_array)
    for pose_id in pose_ids_array:
      pose_data = yoga_poses_data[yoga_poses_data["Pose Id"] == int(pose_id)]
      pose_dict = {
            "Pose Name": str(pose_data['Pose Name'].values[0]),
            "Sanskrit Name": str(pose_data['Sanskrit Name'].values[0]),
            "Difficulty Level":pose_data['Difficulty Level'].values[0],
            "Targeted Muscles":str(pose_data['Targeted Muscles'].values[0]),
            "Targeted Muscles":str(pose_data['Targeted Muscles'].values[0]),
            "Benefits":str(pose_data['Benefits'].values[0]),
            "Instructions":str(pose_data['Instructions'].values[0]),
            "Pre Yoga Instructions":str(pose_data['Pre Yoga Instructions'].values[0]),
            "Post Yoga Instructions":str(pose_data['Post Yoga Instructions'].values[0]),
            "Pose Image":str(pose_data['Pose Image'].values[0]),
        }
      result.append(pose_dict)
    jsonResult = json.dumps(result, ensure_ascii=False)
    return jsonResult
print(str(getPoses("1","1","Headaches")))
fitness_array = ["1","2","3"]
flexibility_array = ["1","2","3"]
outcomes_array = ["Anxiety","Back Pain","Calm","Digestion","Flexibility","Headaches","Asthma","Neck Pain","Sciatica","Pressure"]

# Gradio Interface Integration
iFace = gr.Interface(fn = getPoses,
                     inputs = [
                        gr.Dropdown(fitness_array, label="Select Fitness Level\n 1: Beginner\n2: Intermediate\n3: Advanced"),
                        gr.Dropdown(flexibility_array, label="Select Flexibility Level\n 1: Low\n2: Moderate\n3: High"),
                        gr.Dropdown(outcomes_array, label="Select Desired Outcome")
                        ]
                     ,outputs=gr.JSON())

iFace.launch(share=True, debug=True)

```
